## max return with risk constraint with short selling


---




````python
max_risk =  benchrisk
````

````python


n_stocks = data.shape[1]
reg_lambda = 0.01
cov_matrix_reg = np.cov(returns.T) + reg_lambda * np.eye(n_stocks)
weights = cp.Variable(n_stocks)
port_returns = returns.mean(axis=0) @ weights
risk = cp.quad_form(weights, cov_matrix_reg)

objective = cp.Maximize(port_returns)
constraints = [cp.sum(weights) == 1, risk <= max_risk]

problem = cp.Problem(objective, constraints)
problem.solve()
maxreturnweights = weights.value
````

````python
n = returns.shape[1]
rf = 0.03

weights = cp.Variable(n)

muhat = returns.mean(axis=0) - rf
objective = cp.Minimize(cp.quad_form(weights, np.cov(returns.T) + reg_lambda * np.eye(n_stocks)))


constraints = [muhat@weights == 1,  # sum of weights equals one
              cp.sum(weights) >= 0,
               ]  # weights are non-negative

# Solve the problem
problem = cp.Problem(objective, constraints)
problem.solve()

raw_max_sharpe = weights.value
max_sharpe = raw_max_sharpe/np.linalg.norm(raw_max_sharpe)
# Print the optimal portfolio weights
print(weights.value)
````

## min loss

````python
n_days, n_stocks = returns.shape
weights = cp.Variable(n_stocks)
portfolio_returns = np.mean(returns, axis=0) @ weights


confidence_level = 0.95

expected_shortfall = cp.Variable()

# Calculate es
constraints = [
    cp.sum(weights) == 1,  # Sum of weights should be 1 (full investment)
    cp.sum(cp.pos(expected_shortfall - portfolio_returns)) / n_days <= (1 - confidence_level) * expected_shortfall,
    #portfolio_returns >= 0*benchreturn
]

objective = cp.Minimize(expected_shortfall)

problem = cp.Problem(objective, constraints)
problem.solve()
min_loss_weights = weights.value
min_loss_weights
````

````python
def analyze_cumulative_returns(data):
    data = pd.DataFrame(data)[0]
    # data['cum_returns'] = pd.to_numeric(data['cum_returns'], errors='coerce')
    daily_returns = data.pct_change().dropna()

    # Calculate annualized return
    annualized_return = np.mean(daily_returns) * 252

    # Calculate annualized volatility
    annualized_volatility = np.std(daily_returns) * np.sqrt(252)

    # Calculate Sharpe Ratio
    sharpe_ratio = annualized_return / annualized_volatility

    # Calculate Max Drawdown
    drawdown = data / data.cummax() - 1
    max_drawdown = drawdown.min()

    return annualized_return, annualized_volatility, max_drawdown

max_return_result = analyze_cumulative_returns(maxreturntest1/maxreturntest1[0])
min_risk_result = analyze_cumulative_returns(minrisktest1/minrisktest1[0])
min_var_result = analyze_cumulative_returns(minvartest1/minvartest1[0])
min_loss_result = analyze_cumulative_returns(minlosstest1/minlosstest1[0])


results = pd.DataFrame({'Max return': max_return_result,
                        'Min risk' : min_risk_result,
                        'Min VaR' : min_var_result,
                        'Min loss(CVaR)': min_loss_result},
                       index=['Annualized Return', 'Annualized Volatility', 'Max Drawdown'])
results
````
